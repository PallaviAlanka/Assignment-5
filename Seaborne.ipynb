```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample correlation matrix
correlation_matrix = np.array([[1.0, 0.8, 0.2],
                              [0.8, 1.0, -0.4],
                              [0.2, -0.4, 1.0]])

# Create a DataFrame from the correlation matrix
df = pd.DataFrame(correlation_matrix, columns=['Variable 1', 'Variable 2', 'Variable 3'])

# Create the heatmap
sns.heatmap(df, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

# Add title
plt.title('Correlation Matrix Heatmap')

# Display the heatmap
plt.show()
```

#### lineplot: It is a specific function within seaborn used to create line plots. It takes in two main parameters: the x-axis data and the y-axis data.
#### scatterplot--The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables.
#### heatmap---Using a heatmap to visualise a confusion matrix, time-series movements, temperature changes, correlation matrix and SHAP interaction values.
